```python
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import pandas as pd

iris = load_iris()
df = pd.DataFrame(
    iris.data, 
    columns=iris.feature_names
    )

# Select only "petal length" and "petal width"
X = df[['petal length (cm)', 'petal width (cm)']].values
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=42)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

print(iris.feature_names)

# Scatter plot of the first two features
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=iris.target, cmap='viridis', label=iris.target_names)
plt.xlabel(iris.feature_names[3])
plt.ylabel(iris.feature_names[2])
plt.legend(title="Classes", loc="upper right")
plt.show()
```

```python
clf = tree.DecisionTreeClassifier(max_depth=3, random_state=42)
clf = clf.fit(X, y)

y_pred = clf.predict(X_test)
tain_pred = clf.predict(X_train)

# Evaluate the accuracy of the model
accuracy1 = accuracy_score(y_train, tain_pred)
accuracy2 = accuracy_score(y_test, y_pred)
print("Training Accuracy:", accuracy1)
print("Test Accuracy:", accuracy2)

tree.plot_tree(clf, feature_names=['petal length', 'petal width'], class_names=iris.target_names, filled=True)
plt.show()
```

```python
from sklearn.model_selection import learning_curve
# Utilisation de learning_curve pour gnrer les donnes de la courbe d'apprentissage
train_sizes, train_scores, test_scores = learning_curve(clf, X, y, cv=5, train_sizes=np.linspace(0.1, 1.0, 10))

# Calcul des moyennes et des cart-types des scores
train_mean = np.mean(train_scores, axis=1)
train_std = np.std(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)
test_std = np.std(test_scores, axis=1)

# Trac de la courbe d'apprentissage
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, label='Entranement', marker='o')
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, alpha=0.15)
plt.plot(train_sizes, test_mean, label='Test', marker='o')
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.15)

# Ajout de dtails au graphique
plt.title('Courbe d\'Apprentissage - Arbre de Decision')
plt.xlabel('Taille de l\'Ensemble d\'Entranement')
plt.ylabel('Score (Prcision)')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```

## GrisSearch et Validation Croisee

```python
# Dfinir la grille des hyperparamtres  rechercher
param_grid = {
    'criterion': ['gini','entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [None, 1, 2, 3, 4 , 6, 8, 10, 15, 20, 30],
    'min_samples_split': [2, 3, 4, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Initialiser GridSearchCV avec le classificateur, la grille des hyperparamtres et la validation croise (par exemple, 5-fold)
grid_search = GridSearchCV(clf, param_grid, cv=5, n_jobs=-1, scoring='accuracy')

# Adapter le modle aux donnes
grid_search.fit(X, y)

# Afficher les hyperparamtres optimaux
best_params = grid_search.best_params_
print("Meilleurs hyperparamtres:", grid_search.best_params_)

# Afficher la meilleure prcision obtenue lors de la validation croise
print("Meilleure prcision lors de la validation croise:", grid_search.best_score_)
```
